```python
# Pulled USGS data to DSS via DSSVue plugin.
# Open DSS file and read data into pandas dataframe for analysis.
# All data in UTC timezone.
from pydsstools.heclib.dss import HecDss
from pydsstools.core import TimeSeriesContainer,UNDEFINED
import pandas as pd

dss = HecDss.Open(r"Z:\py\usgs_gage_shp_to_dss\usgs_por_daily_ts.dss")
```

```python
dss.getPathnameList("")
```

```python
# for each SITENO in gdf, get start and end data times
for i,row in gdf.iterrows():
    siteno = row['SITENO']
    # print(siteno)
    # get dss pathname
    pathname_pattern = f"/*/{row['SITENO']}/*/*/*/*/"
    path_pattern_list = dss.getPathnameList(pathname_pattern,sort=1)
    # print(path_pattern_list)
    ts = dss.read_ts(f"/{row['SITENAME']}/{row['SITENO']}/FLOW//1DAY/USGS/")
    
    # Has no flow, check for stage.
    if ts.pytimes is None:
        print (f"{siteno} has no flow data, checking for stage data.")
        ts = dss.read_ts(f"/{row['SITENAME']}/{row['SITENO']}/STAGE//1DAY/USGS {siteno}/")
        
        # Has no Flow or Stage data. Skip.
        if ts.pytimes is None:
            print (f"{siteno} has no flow or stage data, skipping.")
            continue
        
        # Has Stage but no Flow, adding to gdf.  
        else:
            print(f"{siteno} has stage data, but no flow.")
            duration = ts.pytimes[-1] - ts.pytimes[0]  
            duration_in_s = duration.total_seconds()
            print (duration)
            years = divmod(duration_in_s, 31536000)[0]
            gdf.loc[i,'data_start'] = ts.pytimes[0]
            gdf.loc[i,'data_end'] = ts.pytimes[-1]
            gdf.loc[i,'data_years'] = years
    
    # Has Flow, adding to gdf
    else: 
        # print (ts.pytimes[0], ts.pytimes[-1])
        duration = ts.pytimes[-1] - ts.pytimes[0]  
        duration_in_s = duration.total_seconds()
        print (duration)
        years = divmod(duration_in_s, 31536000)[0]
        gdf.loc[i,'data_start'] = ts.pytimes[0]
        gdf.loc[i,'data_end'] = ts.pytimes[-1]
        gdf.loc[i,'data_years'] = years

    # get start and end times for pathname dss record
    # # ts = dss.read_ts(pathname,regular=False,window_flag=0)
    # print(ts.times[0], ts.times[-1])
gdf
```

```python
# Remove null data values from the data_years column.
gdf_dss_data=gdf[gdf.data_years.notnull()]
# Save the geodataframe to a GeoJSON file.
gdf_dss_data.to_file('stage.geojson', EPSG=4326, driver='GeoJSON')
```
